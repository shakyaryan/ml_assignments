# 1What is the average selling price of cars for each dealer, and how does it compare across different dealersE

# 3 What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare

```python
df.columns
```

```python
plt.figure(figsize=(9,5))

sns.histplot(
    data=df,
    x='Price ($)',        # car prices
    hue='Transmission',   # transmission type
    bins=30,
    kde=True
)

plt.title("Distribution of Car Prices by Transmission Type")
plt.xlabel("Car Prices")
plt.ylabel("Count")

plt.show()
```

```python
# IQR calculation for each transmission type
Q1 = df.groupby('Transmission')['Price ($)'].quantile(0.25)
Q3 = df.groupby('Transmission')['Price ($)'].quantile(0.75)

IQR = Q3 - Q1

print("Q1:\n", Q1)
print("\nQ3:\n", Q3)
print("\nIQR:\n", IQR)
```

# 6 How does the average selling price of cars vary by customer gender and annual incomeE

# 12  Is there any seasonal trend in car sales based on the date of sale

```python
df.columns
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Month'] = df['Date'].dt.month
```

```python
monthly_sales = df.groupby('Month').size()
monthly_sales
```

```python
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title("Monthly Car Sales Trend")
plt.xlabel("Month")
plt.ylabel("Number of Cars Sold")
plt.show()
```

# 15 How does the average car price vary across different car models and engine types?

```python
df.groupby(['Model','Engine'])['Price ($)'].mean()
```

```python
df.groupby(['Model','Engine'])['Price ($)'].mean().reset_index()
```

```python
# Top 10 models by sales
top_models = df['Model'].value_counts().head(10).index

plt.figure(figsize=(12,6))
sns.barplot(
    data=df[df['Model'].isin(top_models)],
    x='Model',
    y='Price ($)',
    hue='Engine'
)

plt.xticks(rotation=45)
plt.title("Average Car Price by Top Models and Engine Type")
plt.show()
```

The average car price varies across different car models and engine types. To improve readability, the top-selling models were analyzed. Models with Double Overhead Camshaft engines generally show slightly higher average prices, indicating the influence of engine technology on pricing.
