```python
#What are continuous and categorical variables?
"""
Ans -- >> 

______________________________________

Continuous variables are values that can take any number within a range.

In simple words, they can have decimals and many possible values.

Examples:
Height (170.5 cm, 170.55 cm)
Weight (60.2 kg, 60.25 kg)
Temperature (25.1°C, 25.15°C)
Salary (not fixed steps)
________________________________________________
Categorical variables are values that represent groups or labels.

In simple words, they are categories, not numbers with meaning.

Examples:
Gender (male, female)
Color (red, blue, green)
City (Delhi, Mumbai, Pune)
Yes / N0

"""
```

```python
#How do we handle categorical variables in Machine Learning? What are the 
#common techniques?

"""
Ans: 
________________________________
Label Encoding
Each category is assigned a unique number.
Example:
Low → 0, Medium → 1, High → 2
This is mainly used when the categories have an order.
_______________________________________
One-Hot Encoding
Each category is converted into a separate column with values 0 or 1.
Example:
Color = Red, Blue
Red → [1, 0]
Blue → [0, 1]
This is used when there is no natural order between categories.
__________________________________
Ordinal Encoding
Similar to label encoding, but it is specifically used when categories follow a
meaningful order, like ratings or education level

"""
```

```python
# What do you mean by training and testing a dataset?
"""

Ans:


- Training and testing a dataset means dividing the data into two parts so we can build a
model and then check how good it is.

- Training data is the part of the dataset used to teach the model. The model learns 
patterns, relationships, and parameters from this data. It is like studying from
a textbook.

- Testing data is the part of the dataset used to evaluate the model. The model has never 
seen this data before. It is like giving an exam to see how well the model has learned.

- If the model performs well on training data but poorly on testing data,
it means the model has memorized the data and is not good at generalizing. If it 
performs well on both, the model is considered good.




"""
```

```python
# Why do we have to perform EDA before fitting a model to the data?
"""

Ans : 

- EDA tells us what kind of data we are working with and what problems exist in it.

- EDA helps us:

* identify missing values and decide how to handle them
* detect outliers that can negatively affect the model
* understand the distribution of data (skewed or normal)
* see relationships between features and the target
* check categorical and numerical variables
* find data imbalance problems

If we skip EDA, we may choose the wrong model, use wrong features, or get poor 
results even with a good algorithm.

EDA is like checking ingredients before cooking. If you don’t inspect them, the final 
dish may turn out bad.

So, EDA helps us make better decisions, clean data properly, and build a more accurate
and reliable machine learning model.

"""
```

```python
#What is correlation?
"""
Ans: 

Correlation means how two variables are related to each other.

In simple words, it tells us whether changes in one thing are connected to changes
in another thing.

If both increase or decrease together, they are positively correlated.
If one increases while the other decreases, they are negatively correlated.
If there is no clear relationship, there is no correlation.

For example, as study time increases, exam marks often increase. That shows correlation.
But shoe size and exam marks have no relation, so there is no correlation.

So, correlation helps us understand the strength and direction of the relationship 
between variables.

"""
```

```python
#How can you find correlation between variables in Python?
"""
Ans : 

You can find correlation between variables in Python mainly using Pandas and NumPy.

The most common and simple way is using Pandas.

If your data is in a DataFrame, you can use the corr() function.
It calculates correlation between numerical columns.

In simple words:

* Load data into a DataFrame
* Call .corr()
* It returns a correlation table showing how variables are related

Values range from:

* +1 → strong positive correlation
* −1 → strong negative correlation
* 0 → no correlation

You can also use NumPy to find correlation between two variables using correlation 
functions, but Pandas is preferred for datasets.

So, in Python, correlation is usually found using Pandas’ built-in correlation 
methods on numerical data.

"""
```

```python
#What is causation? Explain difference between correlation and causation with an example.
"""
Ans:
a change in one thing is the reason a change happens in another thing.

_____________________________________
Correlation
Correlation means two variables are related or move together, but one does not necessarily
cause the other.

Causation
Causation means one variable directly affects or produces a change in another variable.
____________________________________________

Simple example:

When ice cream sales go up, drowning cases also go up.
This is correlation because both increase at the same time.

But ice cream does not cause drowning.
The real cause is hot weather. In summer, people eat more ice cream 
and swim more, which increases drowning cases.

"""
```

```python
# What is an Optimizer? What are different types of optimizers? Explain each with an 
# example.

"""
Ans-->> An optimizer is a method used in machine learning to improve a model by reducing its errors.
________________________________________
What does an optimizer do?

During training, the model makes mistakes.
The optimizer checks these mistakes and decides:

how much to change the parameters

in which direction to change them

The goal is to minimize the loss value.

_________________________

The most commonly used optimizers in machine learning and deep learning are these:

1. Stochastic Gradient Descent (SGD)
Used in many basic models.
It updates parameters step by step using small amounts of data.
Simple and effective, but can be slow.

2. Mini-batch Gradient Descent
The most practical version of gradient descent.
Uses small batches of data instead of all data or single points.
Faster and more stable than plain SGD.

3. Adam (Adaptive Moment Estimation)
The most popular optimizer today.
It automatically adjusts learning speed and uses past information.
Works well for most deep learning problems.

4. RMSprop
Commonly used for neural networks and time-series data.
It handles noisy and uneven data well.

In short:
For beginners and most real projects, Adamis usually the first choice.
SGD and Mini-batch are used when more control and simplicity are needed.

"""
```
