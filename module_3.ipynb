Ques1; What is Simple Linear Regression?

Ans: 
SLR finds the relationship between one input (X) and the one input (Y).
its fits a straights line to predict Y from X

e.g. -> predict the house price using area of house
more area-> higher price

eqn -> Y=mX+c

Ques2; Key assumptions of Simple Linear Regression

Ans:
1. relationship should be linear
2. errors should be independent
3. errors should have constant variance
4. errors should be normally distributed
   
   e.g. -> if area increases , price should increase in a straight line pattern
