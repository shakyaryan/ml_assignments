# What is the distribution of discounted prices vs. actual prices

```python
df.columns
```

```python
df['actual_price']
```

```python
df['discounted_price']
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹', "").str.replace(",", "")
```

```python
df['actual_price'].unique()
```

```python
df['actual_price'] = df['actual_price'].astype(int)
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', "").str.replace(",", "")
```

```python
df['discounted_price'] = df['discounted_price'].astype(float)
```

```python
df['discounted_price'].unique()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(np.log1p(df['actual_price']), label='Actual Price', fill=True)
sns.kdeplot(np.log1p(df['discounted_price']), label='Discounted Price', fill=True)

plt.title("Log Distribution of Prices")
plt.legend()
plt.show()
```

Actual and discounted prices show a similar right-skewed distribution, meaning most products fall in the low-to-mid price range with a few expensive outliers.
The discounted price curve is shifted left of the actual price curve, confirming that products are consistently sold at lower prices after discounts.

# 6 What are the most popular product keywords

```python
df.columns
```

```python
import re
from collections import Counter

# 1️⃣ saara text ek string me combine karo
text = " ".join(df['about_product'].astype(str)).lower()

# 2️⃣ symbols hatao
text = re.sub('[^a-zA-Z ]',' ', text)

# 3️⃣ words list banao
words = text.split()

# 4️⃣ useless words hatao (VERY IMPORTANT)
stopwords = ['the','and','for','with','this','that','are','you','your','from',
             'have','has','all','use','can','more','not','will','any','our',
             'out','into','etc','only','get','per','day','new','one']

keywords = [w for w in words if w not in stopwords and len(w) > 3]

# 5️⃣ top keywords nikalo
top_keywords = Counter(keywords).most_common(15)
print(top_keywords)
```

#  8 What is the correlation between discounted_price and rating

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

df.dropna(subset=['discounted_price','rating'], inplace=True)
```

```python
correlation = df['discounted_price'].corr(df['rating'])
print("Correlation:", correlation)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(df['discounted_price'], df['rating'], alpha=0.5)

plt.xlabel("Discounted Price")
plt.ylabel("Rating")
plt.title("Correlation between Discounted Price and Rating")
plt.show()
```
